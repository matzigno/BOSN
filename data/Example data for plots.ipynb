# Example access data for plots

```python
import seaborn as sns
import json
```

```python
sns.__version__
```

## Timeseries and ACF

```python
with open("timeseries.json","r") as f:
    FIGURES_DATA = json.load(f)
```

```python
FIGURES_DATA.keys()
```

### Time series

```python
data = FIGURES_DATA["custom_json_operation"]["custom_json"]
```

```python
sns.lineplot(data["x"],data["y"], legend='brief', label="test" , alpha=0.8 , linewidth=2 )
```

### ACF

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
data = FIGURES_DATA["custom_json_operation_acf"]["custom_json_operation_acf"]
```

```python
plot_acf(data["y"], lags=1000, title="") #), title="{} - ACF".format("STEEM"))
```
